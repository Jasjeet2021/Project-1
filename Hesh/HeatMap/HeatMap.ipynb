```python
# Dependencies
import os
import csv
import pandas as pd
import numpy as np
import requests
import json
import gmaps

# Configure gmaps
gmaps.configure(api_key=gkey)

# Google API Key
from config import gkey
```

```python
# Import cities file as DataFrame
data_file3="Resources/Enrolments.xlsx"
data_file3_df= pd.read_excel(data_file3)
data_file3_df.head()
```

```python
# Add columns for lat, lng, 

data_file3_df["Lat"] = ""
data_file3_df["Lng"] = ""
```

```python
data_file3_df.head()
```

```python
# create a params dict that will be updated with new Area each iteration
params = {"key": gkey}

# Loop through the data_file3_df and run a lat/long search for each Area
for index, row in data_file3_df.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    Area= row['Operational Area']
    state = row['State']
    

    # update address key value
    params['address'] = f"{ Area},{state}"

    # make request
    Area_lat_lng = requests.get(base_url, params=params)
    
    # print the Area_lat_lng url
   
    
    # convert to json
    Area_lat_lng =  Area_lat_lng.json()

    data_file3_df.loc[index, "Lat"] = Area_lat_lng["results"][0]["geometry"]["location"]["lat"]
    data_file3_df.loc[index, "Lng"] = Area_lat_lng["results"][0]["geometry"]["location"]["lng"]

# Visualize to confirm lat lng appear
#data_file3_df.head()
```

```python
# Save Data to csv
data_file3_df.to_csv("Resources/Enrolments_Output.csv")

# Visualize to confirm Operational Area data appears
data_file3_df.head(10)
```

```python
# Store latitude and longitude in locations
locations = data_file3_df[["Lat", "Lng"]]

# Fill NaN values 
Total_Enrolments= data_file3_df["Total Enrolments"]
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=Total_Enrolments, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
